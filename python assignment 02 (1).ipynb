```python
# 6.Hollow Square Pattern 
# Write a program to print a hollow square pattern with `n` rows: Input: n = 5 
for i in range(5):
    for j in range(5):
        if i==0 or i==5-1 or j==0 or j==5-1:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
# 7.Diamond Pattern 
# Write a program to print a diamond pattern with `2n - 1` rows: Input: n = 3 
for i in range(1,4):
    for j in range(1,4-i):
        print(end =" ")
    for j in range(1,i+1):
        print("*",end=" ")
    print()
for i in range(1,4):
    for j in range(1,i+1):
        print(end =" ")
    for j in range(1,4-i):
        print("*",end=" ")
    print()
```

```python
# 8.Pascal's Triangle 
# Write a program to print Pascal's Triangle with `n` rows: Input: n = 5 
n = int(input())
for i in range(n):
    value = 1
    for j in range(n - i - 1):
        print(" ", end="")
    for j in range(i + 1):
        print(value, end=" ")
        value = value * (i - j) // (j + 1)
    print()
```

```python
# 9.Sum of Even Numbers: Write a program that calculates the sum of all even numbers up to `n': 
n=int(input("enter the number"))
sum=0
for i in range(2,n+1,2):
    if i%2==0:
        sum+=i
print(sum)
```

```python
# 10.Factorial Calculation: Write a program to calculate the factorial of a given number `n` using a `for` loop:
n=int(input())
f=1
if n<0:
    print("negative number")
elif n==0 or n==1:
    print("factorial is 1")
else:
    for i in range(1,n+1):
        f*=i
    print(f)
```

```python
# 11.Prime Numbers Up to n: Write a program to print all prime numbers up to `n`:
def prime(n):
    if n>1:
        for i in range(2,n):
            if n%i==0:
                print("not prime")
            else:
                print("prime")
    else:
        print("not prime")
n=int(input())
prime(n)
```

```python
# 12.Vowel Counter: Write a program to count the vowels in a given string using a `for` loop: 
str=input()
str.lower()
c=0
for i in str:
    if i=="a" or i=="e" or i=="i" or i=="o" or i=="u":
        c+=1
print(c)
```

```python
# 13.Reverse a String: Write a program to reverse a given string without using any built-in reverse functions: 
def reverse(s):
    reversed_str=" "
    for i in range(len(s)-1,-1,-1):
        reversed_str+=s[i]
    return reversed_str
s=input("enter the string:")
reversed_s=reverse(s)
print(reversed_s)
```

```python
# 14.Substring Finder: Write a program that finds and prints all substrings of a given string: 
def substring(s):
    substri=[]
    for i in range(len(s)):
        for j in range(i+1,len(s)+1):
            substri.append(s[i:j])
            for substring in substri:
                print(substring)
input_string=input("enter the string:")
substring(input_string)                
```

```python
# 15. Character Frequency Counter: Write a program to print all unique characters and their frequencies in a given string: 
string=input()
char_frequency={}
for char in string:
    if char in char_frequency:
        char_frequency[char]+=1
    else:
        char_frequency[char]=1
for char,freq in char_frequency.items():
    print(f"{char}:{freq}")
```

```python
# 16.Fibonacci Sequence: Write a program that prints the Fibonacci sequence up to `n` terms:
def fib(n):
    a=0
    b=1
    for i in range(0,n):
        print(a,end=" ")
        a,b=b,a+b
fib(10)
```

```python
# 17.Multiplication Table: Write a program that prints a multiplication table for numbers from `1` to `10`: 
for i in range(1,10+1):
    for j in range(1,10+1):
        print(f"{i*j:3}",end=" ")
    print()
```
